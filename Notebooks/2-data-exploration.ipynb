```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
# Cargamos los dataframe con los datos que hemos guardado antes:
data_train = pd.read_csv(r'../Data/processed/train-processed.csv')
data_val = pd.read_csv(r'../Data/processed/val-processed.csv')
```

```python
print('Hay ',len(data_train),' datos para el entrenamiento.')
print('Hay ',len(data_val),' datos para la valoración.')
```

```python
# Histograma de Edades
plt.hist(data_train['Age'], bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Distribución de Edades Train')
plt.grid(True)
plt.show()

# Histograma de Género
plt.hist(data_train['Gender'], bins=2, edgecolor='black', alpha=0.7, density=True)
plt.xlabel('Género (0: Femenino, 1: Masculino)')
plt.ylabel('Frecuencia')
plt.title('Distribución de Género Train')
plt.xticks([0, 1])
plt.grid(True)
plt.show()
```

```python
# Histograma de Edades
plt.hist(data_val['Age'], bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Distribución de Edades Validation')
plt.grid(True)
plt.show()

# Histograma de Género
plt.hist(data_val['Gender'], bins=2, edgecolor='black', alpha=0.7, density=True)
plt.xlabel('Género (0: Femenino, 1: Masculino)')
plt.ylabel('Frecuencia')
plt.title('Distribución de Género Validation')
plt.xticks([0, 1])
plt.grid(True)
plt.show()
```
